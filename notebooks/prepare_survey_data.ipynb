# Load all Data

```python
# Function to load dataset
import pandas as pd
path = "../data/SMILE-College Dataset/smile-college-dataset.csv"
def load_dataset(path):
  df = pd.read_csv(path)
  desired_columns = ['School', 'Comment', 'Labels', 'Validated_Labels']
  df = df[desired_columns]
  return df

data_df = load_dataset(path)
data_df = data_df[data_df['Validated_Labels'] != 'skip']
data_df = data_df.drop(columns=['Labels'])
data_df = data_df.reset_index()
data_df = data_df.drop(columns=['index'])
```

# Process all Data

```python
import numpy as np
np.random.seed(42)
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np
import json
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
x = data_df["Comment"].tolist()
y_ = data_df["Validated_Labels"].tolist()
```

```python
class_mapping = {
    "not-satisfied": 3,
    "satisfied": 1,
    "both": 2,
    "neutral": 0,
}

class_mapping_reverse = {
    3: "not-satisfied",
    1: "satisfied",
    2: "both",
    0: "neutral",
}

y = [class_mapping[class_label] for class_label in y_]

y = np.array(y)

print(y)
```

# Create Splits

```python
import json
import os
from sklearn.model_selection import train_test_split

# Function to split data based on given percentages
def split_data(x, y, train_size=0.7, val_size=0.15, test_size=0.15, random_state=42, stratify=True):
    if stratify:
        stratify_labels = y
    else:
        stratify_labels = None

    # First split: Train set and the remainder (Val + Test)
    x_train, X, y_train, Y = train_test_split(
        x, y,
        test_size=(1.0 - train_size),
        random_state=random_state,
        stratify=stratify_labels
    )

    # Calculate the size ratio of val to test in the remaining data
    val_test_ratio = val_size / (val_size + test_size)

    # Second split: Validation set and Test set
    x_val, x_test, y_val, y_test = train_test_split(
        X, Y,
        test_size=(1.0 - val_test_ratio),
        random_state=random_state,
        stratify=Y if stratify else None
    )

    return x_train, x_val, x_test, y_train, y_val, y_test

x_train, x_val, x_test, y_train, y_val, y_test = split_data(
        x, y,
        train_size=0.75,
        val_size=0.05,
        test_size=0.20,
        random_state=42,
        stratify=True
    )

print(len(y_train))
print(len(y_test))
print(len(y_val))
```

```python
def calculate_lengths(x_train):
    avg_words = sum(len(record.split()) for record in x_train) / len(x_train)
    min_words = min(len(record.split()) for record in x_train)
    max_words = max(len(record.split()) for record in x_train)

    avg_sentences = sum(len(record.split('.')) for record in x_train) / len(x_train)
    min_sentences = min(len(record.split('.')) for record in x_train)
    max_sentences = max(len(record.split('.')) for record in x_train)

    print("AVERAGE WORD - ",avg_words, "MIN WORD - ", min_words, "MAX WORD - ", max_words, "AVERAGE SENTENCE - ", avg_sentences, "MIN SENTENCE - ", min_sentences, "MAX SENTENCE - ", max_sentences)

calculate_lengths(x_train)
calculate_lengths(x_test)
calculate_lengths(x_val)
```

# Save Splits

```python
import json
import os

# Function to create JSON data
def create_json_data(x_data, y_data, class_mapping_reverse):
    data = []
    for text, category in zip(x_data, y_data):
        # label = class_mapping_reverse.get(category, "Unknown")
        label = int(category)
        data.append({"text": text, "label": label})
    return data

# Function to save JSON data to a file
def save_json_data(data, output_folder, file_name):
    os.makedirs(output_folder, exist_ok=True)
    file_path = os.path.join(output_folder, file_name)
    with open(file_path, "w") as f:
        json.dump(data, f, indent=4)
    print(f"Saved {file_name} to {output_folder}")
```

```python
from datetime import datetime

def get_todays_date():
    # Get today's date
    today = datetime.now()
    # Format the date as YYYYMMDD
    formatted_date = today.strftime('%Y%m%d')
    return formatted_date

# Output folder
output_folder = f"../data/splits/{get_todays_date()}/"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Create JSON data for each set
train_data = create_json_data(x_train, y_train, class_mapping_reverse)
val_data = create_json_data(x_val, y_val, class_mapping_reverse)
test_data = create_json_data(x_test, y_test, class_mapping_reverse)

# Save each JSON data to corresponding file
save_json_data(train_data, output_folder, "train.json")
save_json_data(val_data, output_folder, "val.json")
save_json_data(test_data, output_folder, "test.json")

# Save the class mapping and reverse mapping to mapping.json
mapping_data = {
    "class_mapping": class_mapping,
    "class_mapping_reverse": class_mapping_reverse
}
save_json_data(mapping_data, output_folder, "mapping.json")
```

```python
import json
import os

# Function to load JSON from a file
def load_json(folder_path, file_name):
    file_path = os.path.join(folder_path, file_name)
    with open(file_path, "r") as f:
        data = json.load(f)
    return data

# Function to load all JSONs (train, val, test, mapping) into respective variables
def load_all_jsons(folder_path):
    train = load_json(folder_path, "train.json")
    val = load_json(folder_path, "val.json")
    test = load_json(folder_path, "test.json")
    mapping = load_json(folder_path, "mapping.json")

    return train, val, test, mapping

# Example usage:
output_folder = "../data/splits/"
train, val, test, mapping = load_all_jsons(output_folder)
```
